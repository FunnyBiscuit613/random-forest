```python
import os
import rasterio
import numpy as np

from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor 
from sklearn.datasets import load_iris 

def getFileList(dir,Filelist, ext=None):
    """
    获取文件夹及其子文件夹中文件列表
    输入 dir：文件夹根目录
    输入 ext: 扩展名
    返回： 文件路径列表
    """
    newDir = dir
    if os.path.isfile(dir):
        if ext is None:
            Filelist.append(dir)
        else:
            if ext in dir[-3:]:
                Filelist.append(dir)
    
    elif os.path.isdir(dir):
        for s in os.listdir(dir):
            newDir=os.path.join(dir,s)
            getFileList(newDir, Filelist, ext)
 
    return Filelist
```

```python
filelist=getFileList(r'I:\date\各个变量年平均值-海南',[],'tif')
print(filelist)
```

```python
RSM_list = []
for i in filelist:
    if 'RSM' in i :
        RSM_list.append(i)
print(RSM_list)
```

```python
for i in RSM_list:
    dataset = rasterio.open(i)
    band1 = dataset.read(1)
    band1[np.abs(band1)>100]=np.nan
    print(i.split('\\'))
    with rasterio.open(
            'I:\date\临时/{}'.format(i.split('\\')[-1]),
            'w',
              driver='GTiff',
              height=dataset.height,
              width=dataset.width,
              count=1,
              dtype=dataset.dtypes[0],
              crs=dataset.crs,
              transform=dataset.transform,
        ) as dst:
            dst.write(band1, 1)
```
